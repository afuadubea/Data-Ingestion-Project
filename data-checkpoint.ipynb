```python
#function to load data into MySQL workbench
from sqlalchemy import create_engine

def import_files_to_mysql(engine, folder_path):
    #specify your the folder that contains files to import
    folder_path = 'your_folder_path'
    files = os.listdir(folder_path)
    for file_name in files:
        table_name = os.path.splitext(file_name)[0]
        file_path = os.path.join(folder_path, file_name)
        
        #scanning folder for csv files
        if os.path.isfile(file_path):
            try:
                if file_name.endswith('.csv'):
                    df = pd.read_csv(file_path)
                elif file_name.endswith('.json'):
                    df = pd.read_json(file_path)
                else:
                    print(f"Unsupported file format: {file_name}")
                    continue

                df.to_sql(table_name, engine, if_exists='replace', index=False)
                print(f"Table '{table_name}' imported successfully.")
            except Exception as e:
                print(f"Error importing file '{file_name}': {e}")
#defining Mysql user details
def main():
    user_name = 'user_name'
    user_password = 'user_password'
    database = 'database_name'
    host_name = 'host_name'
    folder_path = 'folder_path'

    # Create SQLAlchemy engine
    engine = create_engine(f"mysql+mysqlconnector://{user_name}:{user_password}@{host_name}/{database}")

    import_files_to_mysql(engine, folder_path)

    print("Connection closed")

if __name__ == "__main__":
    main()
```
